```python
from os import path
import pandas as pd
from sklearn.preprocessing  import StandardScaler
from defi_econ.constants import REGRESSION_DATA_PATH
```

```python
data = pd.read_csv(path.join(REGRESSION_DATA_PATH, "regression_data_sample.csv"), index_col = ['symbol', 'date'])
data
```

```python
X = data.loc[:,'asset_price':]
scaler = StandardScaler()
scaler = scaler.fit(X)
X_scaled = scaler.transform(X)
data.loc[:,'asset_price':] = X_scaled
```

```python
data.to_csv(path.join(REGRESSION_DATA_PATH, "regression_data_sample_standard.csv"))
```
