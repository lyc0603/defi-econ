```python
from os import path
import pandas as pd
from defi_econ.constants import BETWEENNESS_DATA_PATH
```

```python
data_v2v3 = pd.read_csv(path.join(BETWEENNESS_DATA_PATH, "betweenness_centrality_merge.csv"))
data_v2v3
```

```python
usdc_value_v2v3 = data_v2v3.loc[data_v2v3['node'] == 'USDC'].set_index(['date'])
weth_value_v2v3 = data_v2v3.loc[data_v2v3['node'] == 'WETH'].set_index(['date'])
wbtc_value_v2v3 = data_v2v3.loc[data_v2v3['node'] == 'WBTC'].set_index(['date'])
usdt_value_v2v3 = data_v2v3.loc[data_v2v3['node'] == 'USDT'].set_index(['date'])
dai_value_v2v3 = data_v2v3.loc[data_v2v3['node'] == 'DAI'].set_index(['date'])
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.figure(figsize=(20,6))
plt.xlabel('Date', fontsize=16)
plt.ylabel('Betweenness Centrality', fontsize=16)
plt.title('Betweenness Centrality (Count Based) by Token', fontsize=20)
plt.plot(usdc_value_v2v3['betweenness_centrality_count'], lw=2, label='USDC')
plt.plot(weth_value_v2v3['betweenness_centrality_count'], lw=2, label='WETH')
plt.plot(wbtc_value_v2v3['betweenness_centrality_count'], lw=2, label='WBTC')
plt.plot(usdt_value_v2v3['betweenness_centrality_count'], lw=2, label='USDT')
plt.plot(dai_value_v2v3['betweenness_centrality_count'], lw=2, label='DAI')
plt.axvline('2021/5/5', c= (.5,.5,.5), ls='--')
plt.legend()
plt.xticks(usdc_value_v2v3.index[::15], rotation = 45)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,6))
plt.xlabel('Date', fontsize=16)
plt.ylabel('Betweenness Centrality', fontsize=16)
plt.title('Betweenness Centrality (Volume Weighted) by Token', fontsize=20)
plt.plot(usdc_value_v2v3['betweenness_centrality_volume'], lw=2, label='USDC')
plt.plot(weth_value_v2v3['betweenness_centrality_volume'], lw=2, label='WETH')
plt.plot(wbtc_value_v2v3['betweenness_centrality_volume'], lw=2, label='WBTC')
plt.plot(usdt_value_v2v3['betweenness_centrality_volume'], lw=2, label='USDT')
plt.plot(dai_value_v2v3['betweenness_centrality_volume'], lw=2, label='DAI')
plt.axvline('2021/5/5', c= (.5,.5,.5), ls='--')
plt.legend()
plt.xticks(usdc_value_v2v3.index[::15], rotation = 45)
plt.show()
```
